```python
URL1 = "https://www.glassdoor.com/Job/palo-alto-data-scientist-jobs-SRCH_IL.0,9_IC1147434_KO10,24_IP"
```

```python
pages = ["https://www.glassdoor.com/Job/jobs.htm?suggestCount=0&suggestChosen=false&clickSource=searchBtn&typedKeyword=Data&locT=C&locId=1147434&jobType=&context=Jobs&sc.keyword=Data+Scientist&dropdown=0"]
for page_number in range(2, 31):
    page_url = URL1 + str(page_number) + ".htm"
    pages.append(page_url)
```

```python
pages[1]
```

```python
data = {'Company': [],
        'Job_Title': [],
        'Location': [],
        'Salary': [], 
        'Job_Sedcription': [],
        'URL': []}

for item in pages:
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}
    page = requests.get(item, headers=headers)
    if page.status_code != 200:
        print('Damm!')
        continue

    
    soup = BeautifulSoup(page.content, features='html.parser')
    jobs_search = soup.find(id='MainCol')
    job_elems = jobs_search.find_all('li', class_="jl react-job-listing gdGrid")
    

    for job_ele in job_elems:

        job_location = job_ele.find('span', class_="loc css-nq3w9f pr-xxsm").text

        c_name, j_title = job_ele.find_all('a')[1], job_ele.find_all('a')[2]

        company_name = c_name.find('span').text

        job_title = j_title.find('span').text

        company_url = 'https://www.glassdoor.com' + job_ele.find('a')['href']


        try:
            job_salary = job_ele.find('div', class_="salaryEstimate css-nq3w9f pr-xxsm").find('span', class_="css-18034rf e1wijj242").text
        except AttributeError:
            job_salary = 'N/A'

        data['Company'].append(company_name)
        data['Job_Title'].append(job_title)
        data['Location'].append(job_location)
        data['Salary'].append(job_salary)
        data['Job_Sedcription'].append('Unavailable ATM')
        data['URL'].append(company_url)
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_csv('Temp_res.csv')
```
